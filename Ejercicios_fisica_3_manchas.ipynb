El archivo `manchasolares.txt`, contiene el número observado de manchas solares en el Sol en cada mes desde enero de 1749. El archivo contiene dos columnas de números, la primera es el mes y la segunda el número de manchas solares.

```python
from matplotlib import pyplot as plt
from pylab import scatter,xlabel,ylabel,xlim,ylim,show
from numpy import loadtxt

data = loadtxt("manchasolares.txt",float)

x = data[:,0]
y = data[:,1]

plt.figure(figsize=(8,5))

scatter(x,y)

plt.title("Manchas solares")

xlabel("Manchas solares", fontsize=15)
ylabel("tiempo", fontsize=15)
xlim(0,3200)
ylim(0,200)

plt.grid()
```

```python
from matplotlib import pyplot as plt
from numpy import loadtxt

data = loadtxt("manchasolares.txt", float)

x = data[:, 0]
y = data[:, 1]

plt.figure(figsize=(8, 5))

# Crear un gráfico de barras
plt.bar(x, y, color='green', width=1.0)

plt.title("Manchas solares")
plt.xlabel("Manchas solares", fontsize=15)
plt.ylabel("Tiempo", fontsize=15)
plt.xlim(0, 3200)
plt.ylim(0, 200)

plt.grid()

plt.show()
```

```python
from matplotlib import pyplot as plt
from numpy import loadtxt

data = loadtxt("manchasolares.txt", float)

x = data[:, 0]
y = data[:, 1]

plt.figure(figsize=(8, 5))

# Crear un gráfico de líneas
plt.plot(x, y, linestyle='-', color='blue')

plt.title("Manchas solares")
plt.xlabel("Manchas solares", fontsize=15)
plt.ylabel("Tiempo", fontsize=15)
plt.xlim(0, 3200)
plt.ylim(0, 200)

plt.grid()

plt.show()
```

```python
from matplotlib import pyplot as plt
from numpy import loadtxt

data = loadtxt("manchasolares.txt", float)

x = data[:, 0]
y = data[:, 1]

plt.figure(figsize=(10, 6))

# Crear un gráfico de líneas sólidas y más ancho
plt.plot(x, y, linestyle='-', linewidth=2, color='navy')

plt.title("Manchas Solares", fontsize=18)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Número de Manchas Solares", fontsize=14)
plt.xlim(1700, 3200)
plt.ylim(0, 200)

# Agregar una cuadrícula
plt.grid(True, linestyle='--', alpha=0.7)

# Agregar una leyenda
plt.legend(["Manchas Solares"], loc='upper right', fontsize=12)

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

data = np.loadtxt("manchasolares.txt", float)

x = data[:, 0]
y = data[:, 1]

# Interpolar los datos para obtener más puntos
x_interp = np.linspace(x.min(), x.max(), 1000)  # Ajusta la cantidad de puntos según tus necesidades
y_interp = np.interp(x_interp, x, y)

plt.figure(figsize=(8, 5))

# Utilizar plot en lugar de scatter para una representación más fluida
plt.plot(x_interp, y_interp, linestyle='-', color='blue')

plt.title("Manchas solares")
plt.xlabel("Manchas solares", fontsize=15)
plt.ylabel("Tiempo", fontsize=15)
plt.xlim(0, 3200)
plt.ylim(0, 200)

plt.grid()

plt.show()

```
