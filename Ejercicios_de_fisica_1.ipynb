# Código de Gas Ideal


```python
## No es necesario editar esta celda, es para que los ejercicios funcionen.

def gasIdeal(N=None,P=None,V=None,T=None):
    R = 8.13446261815324 # J/molK

    if N == None:
        despeje = P*V/(R*T) # Mol

    elif P == None:
        despeje = N*R*T/V # Pa

    elif V == None:
        despeje = N*R*T/P # m^3

    elif T == None:
        despeje = P*V/(R*N) # Kelvin

    return despeje
```

# Ejercicio 2. Despejes de la ecuación de Van der Waals

La ecuación de Van der Waals es
$$
\left(P+\frac{N^2A}{V^2}\right)(V-NB)=NRT
$$
en donde:
* $P$ es la presión en Pascales
* $V$ es el volumen en m$^3$
* $N$ es la cantidad de partículas en Mol
* $R$ es la constante de gas ideal, en J/molK
* $T$ es la temperatura en Kelvin
* $A$ mide la atracción entre partículas (depende de la especie)
* $B$ volumen disponible de un mol de partículas

```python
def ecuacionVanDerWaals(P, T, A, B, N, V):
    # Definir la constante de gas ideal (J/molK)
    R = 8.314  # J/(mol*K)

    # Despejar P y T de la ecuación de Van der Waals
    ecuacion_P = (P + (N**2 * A) / V**2) * (V - N * B) - N * R * T
    ecuacion_T = (P + (N**2 * A) / V**2) * (V - N * B) / (N * R)

    return ecuacion_P, ecuacion_T

# Ejemplo de uso
P_valor = 10  # Presión en Pascales
T_valor = 300  # Temperatura en Kelvin
A_valor = 3.59  # Parámetro A de Van der Waals
B_valor = 0.0427  # Parámetro B de Van der Waals
N_valor = 2  # Cantidad de partículas en mol
V_valor = 0.1  # Volumen en m^3

# Obtener resultados
resultado_P, resultado_T = ecuacionVanDerWaals(P_valor, T_valor, A_valor, B_valor, N_valor, V_valor)

# Imprimir resultados
print(f"Resultado despejado para P: {resultado_P}")
print(f"Resultado despejado para T: {resultado_T}")
```

```python
def despejar_V(P, T, A, B, N):
    # Definir la constante de gas ideal (J/molK)
    R = 8.314  # J/(mol*K)

    # Despejar V de la ecuación de Van der Waals
    # (P + N^2A/V^2)(V - NB) = NRT
    # PV + N^2A/V - PNB - N^2AB/V^2 = NRT
    # PV^2 + N^2A - PNBV - N^2AB = NRTV
    # PV^2 + PNBV + NRTV = N^2A + N^2AB
    # V^2 + (PNB - NRT)V - (N^2A + N^2AB) = 0

    coeficiente_a = 1
    coeficiente_b = P * N * B - N * R * T
    coeficiente_c = -(N**2 * A + N**2 * A * B)

    # Aplicar la fórmula cuadrática para obtener V
    discriminante = coeficiente_b**2 - 4 * coeficiente_a * coeficiente_c
    if discriminante >= 0:
        V_positivo = (-coeficiente_b + discriminante**0.5) / (2 * coeficiente_a)
        V_negativo = (-coeficiente_b - discriminante**0.5) / (2 * coeficiente_a)
        return V_positivo, V_negativo
    else:
        # No hay soluciones reales
        return None, None

# Ejemplo de uso
P_valor = 10  # Presión en Pascales
T_valor = 300  # Temperatura en Kelvin
A_valor = 3.59  # Parámetro A de Van der Waals
B_valor = 0.0427  # Parámetro B de Van der Waals
N_valor = 2  # Cantidad de partículas en mol

# Obtener resultados
resultado_V_positivo, resultado_V_negativo = despejar_V(P_valor, T_valor, A_valor, B_valor, N_valor)

# Imprimir resultados
print(f"Resultado despejado para V (positivo): {resultado_V_positivo}")
print(f"Resultado despejado para V (negativo): {resultado_V_negativo}")
```

```python
moles = 1
volumenes = np.linspace(0,1,500)

temperaturas = 0 # °C

temp = temperaturas+273.15 # Kelvin
presiones = gasVanDerWaals(N=moles,
                               P=None,
                               V=volumenes,
                               T=temp,
                               A=1,
                               B=1)
plt.plot(volumenes,presiones,label=str(temperaturas))

print('imprime el valor de la integral') # <------------- Edita aquí
plt.fill_between(x = volumenes, y1 = presiones) # Comando para colorear


plt.title('Hola mundo')
plt.xlabel('Hola mundo (m$^3$)')
plt.ylabel('Hola mundo (Pa)')

plt.show()
```

```python
import numpy as np
from scipy.integrate import quad

# Parámetros del gas de dióxido de nitrógeno NO2
A_valor = 5.284  # atm L^2/mol^2
B_valor = 0.04424  # L/mol

# Datos de la transformación isotérmica
moles = 6
temperatura_C = -235
temperatura_K = temperatura_C + 273.15
volumen_inicial = 15  # L
volumen_final = 30  # L

# Convertir A y B a las unidades correctas
A_valor = A_valor * 101325  # Convertir a Nm^4/mol^2
B_valor = B_valor  # B ya está en m^3/mol

# Función para la presión en la ecuación de Van der Waals
def presion_van_der_waals(V):
    return (moles * 8.314 * temperatura_K) / (V - moles * B_valor) - (moles**2 * A_valor) / V**2

# Calcular el trabajo usando la integral numérica
trabajo, _ = quad(lambda V: -presion_van_der_waals(V), volumen_inicial, volumen_final)

# Imprimir el resultado del trabajo
print(f"El trabajo realizado durante la expansión isotérmica es: {trabajo} Joules")
```

# Ejercicio 6. Ciclo de Carnot

Para un sistema de 5 mol de gas ideal, con una presión inicial de 1800 Pascales.
1. El ciclo comienza con el gas en 3 m$^3$ y se expande **isotérmicamente** hasta 12 m$^3$.
2. Después se expande **adiabáticamente** hasta 100 Pascales y un volumen final de 15 $m^3$.
3. Tras esto, se comprime **isotérmicamente** hasta un volumen de 5 m$^3$.
4. Finalmente se sigue comprimiendo **adiabáticamente** hasta el volumen y presiones iniciales.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Función para el trabajo en una expansión isotérmica (usando gas ideal)
def trabajo_isotermico(V1, V2, n, R, T):
    return n * R * T * np.log(V2 / V1)

# Función para el trabajo en una expansión adiabática (usando gas ideal)
def trabajo_adiabatico(V1, V2, n, R):
    return n * R * (T1 - T2) / (1 - gamma)

# Función para la presión en una expansión isotérmica (usando gas ideal)
def presion_isotermica(V1, V2, n, R, T):
    return n * R * T / V1

# Condiciones iniciales del sistema
moles = 5
presion_inicial = 1800  # Pascales

# Condiciones iniciales para la expansión isotérmica
V1_1 = 3  # m^3
V2_1 = 12  # m^3
T1 = 300  # Temperatura inicial en Kelvin
R = 8.314  # Constante de gas ideal en J/(mol*K)
gamma = 1.4  # Coeficiente de expansión adiabática para el aire

# Calcular el trabajo y las presiones para la expansión isotérmica
trabajo_1 = trabajo_isotermico(V1_1, V2_1, moles, R, T1)
presion_2 = presion_isotermica(V1_1, V2_1, moles, R, T1)

# Condiciones para la expansión adiabática
V1_2 = V2_1
V2_2 = 15  # m^3
T2 = T1 * (V1_2 / V2_2)**(gamma - 1)  # Temperatura final para la expansión adiabática

# Calcular el trabajo y las presiones para la expansión adiabática
trabajo_2 = trabajo_adiabatico(V1_2, V2_2, moles, R)
presion_3 = presion_inicial * (V1_2 / V2_2)**gamma

# Condiciones para la compresión isotérmica
V1_3 = V2_2
V2_3 = 5  # m^3

# Calcular el trabajo y las presiones para la compresión isotérmica
trabajo_3 = trabajo_isotermico(V1_3, V2_3, moles, R, T1)
presion_4 = presion_isotermica(V1_3, V2_3, moles, R, T1)

# Condiciones para la compresión adiabática
V1_4 = V2_3
V2_4 = V1_1

# Calcular el trabajo y las presiones para la compresión adiabática
trabajo_4 = trabajo_adiabatico(V1_4, V2_4, moles, R)

# Graficar el ciclo de Carnot
volumenes = [V1_1, V2_1, V2_2, V2_3, V1_1]
presiones = [presion_inicial, presion_2, presion_3, presion_4, presion_inicial]

# Configurar la gráfica
plt.plot(volumenes, presiones, marker='o', linestyle='-', color='b')
plt.title('Ciclo de Carnot para un Gas Ideal')
plt.xlabel('Volumen (m^3)')
plt.ylabel('Presión (Pascales)')
plt.grid(True)
plt.show()
```
